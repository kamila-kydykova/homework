````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def variance(values): 
    average = mean(values)
    my_sum = 0
    for num in values: 
        my_sum += (num - average)**2
    return my_sum/len(values)
        
````

**Concept:** *Accumulator pattern with numeric result, list iteration, index-based loop.*

**Task:**

Write a function `covariance(xs, ys)` that computes the population covariance between two series:

$$
\text{covariance} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})
$$

Rules:

* If `xs` and `ys` are not the same length, print:
    
    `xs and ys must be the same length`
    
    and return `0`.

* Must use an index-based loop.

* Must use your `mean()` function.

* Then run the given test case and print exactly in the same format.

**Sample Output:**

```
x values: [6, 3, 7, 9, 5]
y values: [4, 1, 7, 5, 0]
covariance: 3.8
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def covariance(xs, ys): 
    x_mean = mean(xs)
    y_mean = mean(ys)
    step_three = [(x - x_mean) * (y - y_mean) for x,y in zip(xs, ys)]
    step_four = sum(step_three)/len(step_three)
    return step_four
````

**Problem 12**

**Concept:** *Accumulator pattern to process a string, produce a string result.*

**Task:**

Write a function `cipher(s, n)` that applies a Caesar cipher shift to a plaintext string.

Rules:

* Assume lowercase letters only for shifting (`a` to `z`).

* Do not shift non-letters (spaces stay spaces, punctuation stays punctuation).

* Must wrap around the alphabet if shifting past `z` or before `a`.

* Must use an accumulator string.

* Then run the exact test cases shown and print in the same format.

**Sample Output:**

```
plaintext = abc, ciphered = def
plaintext = abc xyz, ciphered = def abc
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def cipher(s, n):
  new_str = ""
  for x in s:
    if ord(x) < 97 or ord(x) > 122:
      new_str += x
      continue
    number = ord(x) + n
    if number > 122:
      diff = number - 122
      number = 97 + diff
    elif number < 97:
      number += 26
    new_char = chr(number)
    new_str = new_str + new_char
  return new_str

````

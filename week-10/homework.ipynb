**Review Problem 2**

**Concept:** *reading CSV data, matplotlib, line graph*

**Task:**

Write a function `read_and_plot_column(filename, column_name)` that:
* Opens a CSV file (do **not** use pandas)
* Extracts the values from the specified column
* Casts all values to `float`
* Plots the values using `plt.plot()`
* Adds a title
* Displays the plot

````python
# Your Solution

def read_and_plot_column(filename, column_name):
    """
    Read a CSV file, extract a column as floats, and plot it.
    """
    with open(filename, 'r') as f:
        lines = f.readlines()
    
    headers = lines[0].strip().split(',')
    data = {h: [] for h in headers}

    for line in lines[1:]:
        values = line.strip().split(',')
        for i, h in enumerate (headers):
            data[h].append (values[i])

    for h in data:
        data[h] = np.array(data[h])

    values = [float(v) for v in data[column_name]]

    plt.plot(values)

    plt.xlabel("x values")
    plt.ylabel(column_name)
    plt.title(f"{filename}")

````

**Task:**

Using `df_weather`:
* Write a helper function to extract time (in hours) from a datetime string and convert it to decimal hours (float)
* Convert time using:

    `hours + (minutes / 60) + (seconds / 3600)`
* Create a new DataFrame with the same index as `df_weather`
* Add the following columns:
  * `sunrise` (converted to float)
  * `sunset` (converted to float)
  * `tempmin` (copied directly from `df_weather`)
* Plot all columns together using `.plot()`


**Sample Output:**

This is similar to how your new dataframe should look before plotting:

```
   start_time  end_time  price
P1       8.25     17.50     20
P2       8.50     17.75     35
P3       9.00     18.00     15
P4       9.25     18.25     60
P5       9.50     18.50     50
```

````python
# Your Solution

def float_values(datetime_string):
  """
  Convert a datetime string to decimal hours. 
  """
  datetime = pd.to_datetime(datetime_string)
  
  hours = float(datetime.hour)
  hours += datetime.minute/60 
  hours += datetime.second/3600
  
  return hours
  
dataframe = pd.DataFrame(index=df_weather.index)

dataframe['sunrise'] = df_weather['sunrise'].apply(float_values)
dataframe['sunset'] = df_weather['sunset'].apply(float_values)

dataframe['tempmin'] = df_weather['tempmin']

dataframe.plot()
plt.title("Sunrise, Sunset, and TempMin Values")
plt.show()

````

**Task:**


Write a function `filter_rows_matching_pattern(df, column, pattern)` that:
* Returns rows where the specified column contains the given pattern

* Prints the number of matching rows

**Sample Output:**

If your selected column contained matching text values, the following could be your output:

````python
# Your Solution

def filter_rows_matching_pattern(df, column, pattern):
    """
    Return rows where the given column contains the pattern.
    Handles missing values and prints the number of matching rows.
    """
    filtered_rows = df[df[column].str.contains(pattern)]
    values_left = len(filtered_rows)
    print("Found " + str(values_left) + " records of " + str(column) + " containing " + str(pattern) + ".")
    return filtered_rows
````


````python
# Your Solution

def create_df_categorical(df_source, column_name): 
  """
  Create and return a dataframe with counts and percentages for each category in a column. 
  """
  column_values = df_source[column_name]
  counts = column_values.value_counts()
  percentage_counts = column_values.value_counts(normalize=True)
  new_dataframe = pd.DataFrame()
  new_dataframe["count"] = counts
  new_dataframe["pct"] = percentage_counts
  return new_dataframe

````

**Problem 17**

**Concept:** *Filtering records by category.*


**Task:**

Write a function `filter_weather(df, conditions, temp_min, temp_max)` that:
* Filters rows where:
  * conditions contain the given string
  * `tempmax` ≤ `temp_max`
  * `tempmin` ≥ `temp_min`
* Prints intermediate counts
* Returns the filtered DataFrame

**Sample Output:**

If your data matched the filtering conditions step by step, the following could be your output:
```
4 records with conditions Blue
3 records with conditions Blue and tempmax <= 50
2 records with conditions Blue and 20 <= temp <= 50
Found 2 records with conditions of Blue and temp between 20 and 50.
```

````python
# Your Solution

def filter_weather(df, conditions, temp_min, temp_max):
    """
    Filter rows by conditions and temperature range, print intermediate counts, and return the filtered dataframe. 
    """
    filtered_dataframe = df[df["conditions"].str.contains(conditions)]
    values_left = len(filtered_dataframe)
    print(str(values_left) + " records with conditions " + str(conditions))

    filtered_dataframe = filtered_dataframe[filtered_dataframe["tempmax"] <= temp_max]
    values_left = len(filtered_dataframe)
    print(str(values_left) + " records with conditions " + str(conditions) + " and tempmax <= " + str(temp_max))

    filtered_dataframe = filtered_dataframe[filtered_dataframe["tempmin"] >= temp_min]
    values_left = len(filtered_dataframe)
    print(str(values_left) + " records with conditions " + str(conditions) + " and " + str(temp_min) + " <= temp <= " + str(temp_max))
    print("Found " + str(values_left) + " records with conditions of " + str(conditions) + " and temp between " + str(temp_min) + " and " + str(temp_max) + ".")
    
    return filtered_dataframe

    
````

**Concept:** *Creating DataFrame, standardizing values.*


**Task:**

Write a function `create_standardized_values(df_source, columns, row_index=None)` that:
* Creates a new DataFrame
* Standardizes values relative to a given row
* Uses the first row if none is provided
* Returns the new DataFrame

**Sample Output:**

If the following was your dataframe:
```
      high_value  low_value
2020        100        50
2021        120        55
2022        140        60
2023        160        70
2024        180        75
```
If you call the function
`create_standardized_values(df_example, ["high_value", "low_value"], 2020)`,
you should get the following as the output:
```
      high_value  low_value
2020         1.0       1.00
2021         1.2       1.10
2022         1.4       1.20
2023         1.6       1.40
2024         1.8       1.50

```

````python
# Your Solution

def create_standardized_values(df_source, columns, row_index=None):
    """
    Create and return a new DataFrame with selected columns standardized
    relative to a given row. If row_index is None, use the first row.
    """
    if row_index is None:
        base_row = df_source.iloc[0]
    else:
        base_row = df_source.loc[row_index]

    df_new = df_source[columns].copy()

    for col in columns:
        df_new[col] = df_new[col] / base_row[col]

    return df_new
````
